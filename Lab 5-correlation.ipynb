```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv(r'current-power.csv', delimiter=',')
df.head()
```

```python
df.describe()
```

```python
df.cov()
```

```python
x = df['Current']
y = df['Power']

plt.Figure(figsize=(10,6))
plt.scatter(x, y)
plt.title('Device Current vs Power Response')
plt.xlabel('Current (A)')
plt.ylabel('Power (W)')
plt.show()
```

```python
df.corr()
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept

plt.figure(figsize=(10,6))
plt.scatter(x, y,
              color='darkblue',
              alpha = 0.6,
              label='Data Points')

plt.plot(x, line,
         color='blue',
         label=f'Fitted Line (r = {r_value:.3f})')

plt.title('Device Current vs Power Response')
plt.xlabel('Current (A)')
plt.ylabel('Power (W)')
plt.legend()
plt.show()
```

## NOx Humidity

```python
air = pd.read_csv(r'Statistics_CSV/air-quality.csv', delimiter=',')
air
```

```python
x = air['NOx']
y = air['Relative_Humidity']

slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept

plt.figure(figsize=(10,6))
plt.scatter(x, y,
              color='blue',
              alpha = 0.2,
              label='Data Points')

plt.plot(x, line,
         color='lightblue',
         label=f'Regression Line (r = {r_value:.3f})')

plt.title('NOx vs Relative Humidity')
plt.xlabel('NOx (ug/cu.m)')
plt.ylabel('Relative Humidity (%)')
plt.legend()
plt.show()
```

## NO2 Humidity

```python
x = air['NO2']
y = air['Relative_Humidity']

slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept

plt.figure(figsize=(10,6))
plt.scatter(x, y,
              color='red',
              alpha = 0.2,
              label='Data Points')

plt.plot(x, line,
         color='darkblue',
         label=f'Regression Line (r = {r_value:.3f})')

plt.title('NO2 vs Relative Humidity')
plt.xlabel('NO2 (ug/cu.m)')
plt.ylabel('Relative Humidity (%)')
plt.legend()
plt.show()
```
