```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv(r"transformer-voltage-10-sample.csv", delimiter = ",", index_col='Sample')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
alpha = 0.05
n = df['Voltage'].count()

t_critical = stats.t.ppf(1 - alpha/2, n-1)
t_critical
```

```python
mean = df['Voltage'].mean()
std = df['Voltage'].std(ddof=1)

ci_lower, ci_upper = stats.t.interval(
    confidence = 0.95,
    df = n-1,
    loc = mean,
    scale = std / np.sqrt(n)
    )

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
mean = df['Voltage'].mean()
std = df['Voltage'].std(ddof=1)

x_min = df['Voltage'].min()
x_max = df['Voltage'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)
plt.plot (x,p,
          linewidth=2,
          color='orange')

plt.hist(df['Voltage'],
         bins=5,
         density=True,
         alpha=0.8,
         color='darkblue')

plt.title('Voltage Distribution', font = 'Times New Roman', fontsize=12)
plt.xlabel('Voltage')
plt.ylabel('Probability Density')

plt.axvline(x = ci_lower, color='orange', linestyle='--', label='95% CI Lower Bound')
plt.axvline(x = ci_upper, color='orange', linestyle='--', label='95% CI Upper Bound')
plt.show()
```

# The Shoe Invetory

```python
df = pd.read_csv(r"Statistics_CSV\shoe-inventory.csv", delimiter = ',')
df
```

```python
df.shape
```

```python
df.describe()
```

```python
df.shape
```

```python
mean_size_6 = df.iloc[0,1:13].mean()
mean_size_6
```

```python
std_size_6 = df.iloc[0,1:13].std()
std_size_6
```

```python
n = df.shape[1] - 1

lower_6, upper_6 = stats.t.interval(
    confidence = 0.95,
    df = n - 1,
    loc = mean_size_6,
    scale = std_size_6 / np.sqrt(n)
)

print(f"95% Confidence Interval for Size 6 Shoes: ({lower_6:.2f}, {upper_6:.2f})")
```

```python
mean = df.iloc[0:17,1:13].mean(axis = 1)
std = df.iloc[0:17,1:13].std(axis = 1)

df['ci_lower'], df['ci_upper'] = stats.t.interval(
    confidence = 0.95,
    df = n - 1,
    loc = mean,
    scale = std / np.sqrt(n)
)
```

```python
df
```

```python
df['ci_lower'] = df['ci_lower'].fillna(0)
df['ci_upper'] = df['ci_upper'].fillna(0)
```

```python
df['ci_lower'] = df['ci_lower'].astype(int).round(2)
df['ci_upper'] = df['ci_upper'].astype(int).round(2)
```
